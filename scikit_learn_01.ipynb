# Concepts of Splitting The DATA

```python
from sklearn.datasets import load_iris # Loading and inspecting  the data 
from sklearn.model_selection import train_test_split
```

When building a Machine Learning model, you should never test your model on the same data it learned from. That would be like giving a student the exact questions and answers to study, and then giving them the exact same test—they would get 100% just by memorizing, not by actually understanding!

Training Set (~80%): Used by the algorithm to learn patterns.

Testing Set (~20%): Kept hidden from the algorithm during training. It is used at the very end to evaluate how well the model generalizes to new, unseen data.

# StratifiedShuffleSplit


What does StratifiedShuffleSplit do?

It preserves the same class proportions in both the training and testing sets.

i.e. the distribution percentage among training and testing the percentages stay almost the same.

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1,test_size=0.2)


for train_idx,test_idx in split.split(X,y):
    X_train , X_test = X[train_idx],X[test_idx]
    y_train , y_test = y[train_idx],y[test_idx]
```

# Preprocessing 

Preprocessing converts raw, inconsistent data into a clean and suitable format for machine learning algorithms. It improves model performance, ensures fair comparison between features, and helps algorithms learn meaningful patterns from the data like (0,1,2...
)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# StandardScaler
standardScaler is a preprocessing class available in the sklearn.preprocessing module.


Its purpose is to standardize numerical features so that every feature has:


Mean = 0.     Standard Deviation = 1

```python
X,y = load_iris(return_X_y=True)
```

```python
X
# the features with larger values (5.1,5.7) can dominate some (0.2) machine learning algorithms.

# StandardScaler brings all features to a similar scale
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled =scaler.transform(X_test)
```

```python
X_train_scaled
# now we can see that all the data are roughly on the same scale 
```

```python
# Another method by using numpy 

import numpy as np 
((X_train - np.mean(X_train,axis=0)) / np.std(X_train,axis=0))
```
