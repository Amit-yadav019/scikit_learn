What does StratifiedShuffleSplit do?

It preserves the same class proportions in both the training and testing sets.

i.e. the distribution percentage among training and testing the percentages stay almost the same.

Calculates the mean and standard deviation for each feature using only the training data.

Applies the standardization formula to every training sample:  z = (x−μ) / σ .

Why only transform()?

The test data should be transformed using the same mean and standard deviation learned from the training data.

We do not call fit() again because that would calculate new statistics from the test set and introduce data leakage.

# clustering 

Clustering is an unsupervised machine learning technique used to group similar data points together based on their characteristics

In simple words:

Clustering automatically discovers natural groups in data without using labels.

```python
from sklearn.datasets import make_blobs
# make_blobs() is a Scikit-learn function used to generate an artificial (synthetic) dataset 
# for clustering and classification experiments.

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
X,_ = make_blobs(n_samples=5000,centers=5,random_state=10)
# make_blobs() returns two outputs. = Features (X) and True cluster labels (y). 
# Then why use _?
# Because in clustering, we usually don't need the true labels. We only need the feature values.

# The underscore (_) is a Python convention meaning: "I received this value, but I'm intentionally ignoring it."


# n_samples = 5000 : Specifies the total number of 5000 data points to generate
# centers = 5 : 5 numbers of clustors or groups are created 
# random_state = 10 : Controls the randomness of the generated data
      #. Without fixing random_state, every run creates a different dataset
X
```

```python
X.shape

 # as 5000 samples and bydefault 2 features 
```

```python
scaler = StandardScaler()
# At this point:
# No data has been analyzed.
# No scaling has happened.
# The object is simply ready to perform standardization.

X_scaled = scaler.fit_transform(X)
# This performs two operations in one line

# Part 1: fit(X)
# The scaler first studies the dataset.For every feature (column), it calculates: Mean (μ) and Standard Deviation (σ)

# Part 2: transform(X)
# Now the scaler converts every value using the z-score formula:z= (x−μ) /σ
# where: x = original value , μ = mean & σ = standard deviation


# This transforms each feature so that it has:
# Mean ≈ 0 & Standard Deviation ≈ 1

```

```python
import matplotlib.pyplot as plt 
plt.scatter(X_scaled[:,0],X_scaled[:,1]) # for features 1 and 2 

# : means select all rows and 0 means column this result [:,0] means select all rows and 0 column 

# and  we are plotting only 0 and 1 becoz we have not mention n_features and by default it is 2 
```

Why use a scatter plot?

In clustering, we want to see whether the data naturally forms groups.

Scatter plots make these groups easy to observe.




```python
from sklearn.cluster import KMeans # to create clustering module 
K_means = KMeans(n_clusters=5) # This line creates a K-Means model object.
# At this stage:
# No learning has happened.
# No clusters have been formed.
# The model is simply initialized with the settings you provided.


# Train the Model
K_means.fit(X_scaled)
# fit() means:Train the model using the given data. For K-Means, training means discovering clusters in the dataset.
# and its only .fit(X_scaled) not y terms becoz It does not need target labels (y)
```

```python

plt.scatter(X_scaled[:,0],X_scaled[:,1],c=K_means.labels_) 
# c = , c stand for color 
```

# PCA 
Principal Component Analysis (PCA) is a dimensionality reduction technique used in Machine Learning to reduce the number of 
features (columns) in a dataset while preserving as much important information (variance) as possible.


PCA transforms a high-dimensional dataset into a lower-dimensional dataset without losing much of the important information.



```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X, y = fetch_openml('mnist_784', return_X_y=True)
# X → Feature matrix (input data)
# y → Target labels (output)

# MNIST = Modified National Institute of Standards and Technology database.
# 784 = Each image contains (28*28) 784 pixels
# and each pixels represents 1 features i.e. total 784 features 



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
pca = PCA(n_components=10)

X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
X_train.shape
```

```python
X_train_reduced.shape # as we can see the features is reduced 
```

```python
clf = LogisticRegression(max_iter=100)
# max_iter = 100 means Allow the algorithm to perform at most 100 iterations(no. of times) while finding the best parameters.
# if find before then stop , model learn by imporving step wise step as there max_iter = 100 


clf.fit(X_train, y_train) # The model learns the relationship between Input Image (X) - Correct Digit (y)

print(clf.score(X_test, y_test))

# before reducing features  we got 91% accuracy 
```

```python
clf = LogisticRegression(max_iter=100)
clf.fit(X_train_reduced, y_train)

print(clf.score(X_test_reduced, y_test))

# after reducing features we still got accurasy = 80%

# similarlly if we increase the n_component the accurracy will be increase 
```

# Metrics

In Machine Learning, metrics are quantitative measures used to evaluate how well a machine learning model performs. They compare the model's predictions with the actual (true) values and provide a numerical score that indicates the model's performance.

Different machine learning tasks require different evaluation metrics.


For example:


Classification Metrics:

Accuracy

Precision

Recall

F1-Score

ROC-AUC





Regression Metrics:


Mean Absolute Error (MAE)

Mean Squared Error (MSE)

Root Mean Squared Error (RMSE)

R² Score







Classification Metrics


1. Accuracy

Accuracy measures how many predictions the model made correctly out of all predictions.



2. Precision

Precision measures how many of the predicted positive cases are actually positive.



3. Recall


Recall measures how many actual positive cases the model correctly identified.


4. F1-Score

F1-Score combines Precision and Recall into a single measure to evaluate the model's performance.



5. ROC-AUC

ROC-AUC measures how well the model can distinguish between different classes.

Regression Metrics

1. Mean Absolute Error (MAE)

MAE measures the average difference between the actual values and the predicted values.



2. Mean Squared Error (MSE)

MSE measures the average squared difference between the actual values and the predicted values. Larger errors are penalized more.



3. Root Mean Squared Error (RMSE)

RMSE measures the average prediction error in the same units as the target variable.



4. R² Score

R² Score measures how well the model explains the variation in the data.

Cross Validation is a technique used to evaluate a machine learning model by training and testing it on different parts of the dataset multiple times. It provides a more reliable estimate of the model's performance than using a single train-test split.


Cross Validation checks a model's performance by training and testing it on different parts of the data multiple times.
